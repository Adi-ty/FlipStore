## The Dataset contains the purchase history of a user

```python
purchases = pd.read_csv('book1.csv', header = None)
purchases.head()
```

### Removing the 'nan'

```python
purchaseList = []
for i in range(len(purchases)):
    purchaseList.append([str(purchases.values[i,j]) for j in range(0, 10) if str(purchases.values[i,j])!='nan'])
purchaseList
```

## Apriori Algorithm for Market Basket Analysis

```python
rules = apriori(purchaseList, min_support = 0.001, min_confidence = 0.03, min_lift = 3, min_length = 1)
```

```python
a = 'Denim Shirt'
b = 'Anarkali kurti'
sugg = []
final = []
i = 0
for item in rules:
    i+=1
    pair = item[0]
    items = [x for x in pair]
#     print(items)
    if a in items:
#         print(items)
        sugg = [x for x in items]
        final+=sugg
#         print(final)
final = list(set(final))
print("hi ",final)
```

```python
j =0
for item in rules:
    print(type(item))
    # First index of the inner list contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    to_print = "Rule: "

    arrow = " -> "
    for i in range(len(items)):
        to_print += str(items[i]) + arrow

    print(to_print)
    # Print the support for this association rule
    print("Support: " + str(item[1]))
    # Print the confidence and lift for this association rule
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
    j+=1
print("\n\n", j)
```
